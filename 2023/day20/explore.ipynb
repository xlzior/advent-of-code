```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
with open('./puzzle.in', 'r') as f:
  data = f.read()

for line in data.split("\n"):
  n, o = line.split(" -> ")
  name = n.strip("%&")
  out = o.split(", ")
  for o in out:
    G.add_edge(name, o)
```
